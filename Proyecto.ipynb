```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
df = pd.read_csv('forbes-dataset-2020.csv')
df
```

```python
df1 = df.copy()
df1.drop('S.NO',axis=1,inplace=True)

# Convert string to datetime64
df1['Year'] = df1['Year'].apply(pd.to_datetime,format='%Y')

#Set Date column as the index column.
df1['year'] = pd.DatetimeIndex(df1['Year']).year  
df1.set_index('year', inplace=True)
df1.drop('Year',axis=1,inplace=True)

# Converting the sport column to uppercase
df1['Sport'] = df1['Sport'].str.upper()
df1

# df is the original dataframe while df1 is a copy where the Date has been set as an index column
```

```python
#Deportistas cuyo ranking ha subido al menos dos lugares entre 2010 y 2020.
df2 = df1[df1.index > 2009].copy() #se hace una copia del df1 para tener del 2009 en adelante
df2
```

```python
df2['Previous Year Rank'] = df2['Previous Year Rank'].str.strip('>') #se quita el simbolo > de Previous Year Rank	
df2
```

```python
dfa = df2['Name'].value_counts().to_frame() #se ceuntan las veces que sale cada nombre de un jugador para no tenerlos repetidos
lista_nomb1 = dfa.index[dfa['Name']>1].tolist() #se pasan a una lista los nombres 
```

```python
df2['aparicion'] = df2['Name'].apply(lambda x: 1 if (x in lista_nomb1) else 0) #1 es sale mas de una vez, 0 es sale una vez
df2
```

```python
dfb = df2[df2['aparicion'] == 0] #comenzando con lo que salen una sola vez
dfb
```

```python
dfb = dfb[dfb['Previous Year Rank'].astype(str).apply(lambda x: '?' not in x)] #quitar los que tengan el simbolo ?
dfb = dfb.dropna() #quitar los nan
dfb
```

```python
dfb['Previous Year Rank'] = dfb['Previous Year Rank'].astype(int) #se ponen como entero la columna Previous Year Rank
dfc = dfb[['Name','Current Rank','Previous Year Rank']] #se crea otro df con solo tres columna ('Name','Current Rank','Previous Year Rank')
dfd = dfc[dfc['Current Rank'] + 2 <= dfc['Previous Year Rank']] #se compara el año anterior con el actual si subió al menos 2 lugares en ranking
dfd
```

```python
dfd = dfd.set_index('Name') #se configura el index con los nombres de los jugadores
dfd = dfd.reindex(columns=['Previous Year Rank','Current Rank'])
dfd = dfd.rename(columns={"Previous Year Rank": "Max_Rank", "Current Rank": "Min_Rank"}) #se le da formato al df
dfd
```

```python
df3 = df2.pivot_table(index= ['Name'], columns = 'year', values='Current Rank') #ahora para los que salen mas de una vez, se crea otro df pero ahora con los nombres como indice y año como columnas
df4 = df3.transpose() #transponer el df anterior para mejor manejo, ahora nombres son columnas e index las fechas
df4
```

```python
listaMax = [] #se crean listas para el mayor y menor ranking y para los nombres
listaMin = []
listaNom = []
for i in df4.columns[:]:  #se recorre cada nombre(columna) y obtengo el min y max de las mismas y se guardan en las listas enteriores
  listaMin.append(df4[i].min(skipna = True))
  listaMax.append(df4[i].max(skipna = True))
  listaNom.append(i)

df5 = pd.DataFrame(list(zip(listaMax,listaMin)), index = df3.index,columns = ['Max_Rank','Min_Rank']) #se crea un df con las listas y datos anteriores
dffinal = df5[df5['Max_Rank'] >= df5['Min_Rank'] + 2] #se compara y obtienen aquellos que subieron de ranking al menos 2 posiciones
dffinal
```

```python

#RESPUESTA PUNTO 1

pd.concat([dfd,dffinal]) #se concatenan los resultados obtenidos para tener una sola tablas los jugadores que subieron de ranking al menos 2 posiciones
```

```python
# 2. Atleta con el menor número de apariciones y mayores ganancias.


#df3.index[:]
df6 = df1.groupby(['Name']).sum() #se agrupa por nombre y se suman las ganancias de cada uno
df6.drop('Current Rank',axis=1,inplace=True) #se borra una columna extra
df6
```

```python
df7 = df1['Name'].value_counts().to_frame() #se cuentan las veces que aparece un jugador y lo convierto en df
df7.rename(columns={'Name':'Apariciones'},inplace=True) #se renombra una columna
df7
```

```python
df8 = pd.concat([df6,df7], axis = 1 ) #se concatena las ganancia con las apariciones
df8
```

```python

#RESPUESTA PUNTO 2

df9 = df8[df8['Apariciones'] == 1] #se buscan los jugadores que hayan aparecido una vez y se guardan en un df
gmax = df9['earnings ($ million)'].max() #se calcula el valor de ganancias maximo
df9[df9['earnings ($ million)']==gmax] #se busca quien ganó esa cantidad
```

```python
# 3. Deporte y país con mayor número de atletas no rankeados que entraron en la lista de atletas mejor pagados.


df10 = df1['Nationality'].value_counts().to_frame() #se cuentan las veces que sale un solo pais para no tener repetidos
menor_natio = df10['Nationality'].min() # se calcula se pais aparece menos veces

#df11 = df10[df10['Nationality'] == menor_natio]

lista_paises = df10.index[df10['Nationality'] == menor_natio].tolist() #se guardan en una lista los paises que aparecen con el menor numero de apariciones
lista_paises
```

```python
df12 = df1['Sport'].value_counts().to_frame() #se cuentan las veces que aprecen los deportes para no repetir
menor_sport = df12['Sport'].min() #se calcula el mejor numero de apariciones de los deportes

#df13 = df12[df12['Sport'] == menor_sport]
lista_sport = df12.index[df12['Sport'] == menor_sport].tolist() #se guarda en una lista los deportes que salen el menor numero de veces
lista_sport
```

```python
df1['dep'] = df1['Sport'].apply(lambda x: 1 if (x in lista_sport) else 0) #del df1 se agrega una columna 'dep' para poner a 1 si se está en la lista de deportes con menos apariciones, de lo contrario es 0
df1['pais'] = df1['Nationality'].apply(lambda x: 1 if (x in lista_paises) else 0) ##del df1 se agrega una columna 'pais' para poner a 1 si se está en la lista de paises con menos apariciones, de lo contrario es 0
```

```python

#RESPUESTA PUNTO 3

dfsp = df1[(df1['dep']==1) & (df1['pais']==1)] #se filtra para tener al jugador que pertenece a la lista de paises y al mismo tiempo al deporte con menos apariciones
dfsp[['Sport','Nationality']]
```
